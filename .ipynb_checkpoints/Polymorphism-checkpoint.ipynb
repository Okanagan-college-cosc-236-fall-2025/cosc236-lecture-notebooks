```java
class Animal {
    // private final String name;

    public Animal(String name) { this.name = name; }
    public String getName() { return this.name; }

    // Base behavior (can be overridden)
    public String speak() { return "..."; }

    public String introduce() {
        return "I'm " + this.name + " and I say: " + speak();
    }
}
```

```java
class Dog extends Animal {
    
    public Dog(String name) { 
        super(name);
    }
    @Override
    public String speak() {
        return "woof"; 
    }
}

```

```java
class Cat extends Animal {
    public Cat(String name) 
    { 
        super(name);
    }
    @Override 
    public String speak()
    { 
        return "meow"; 
    }
}
```

```java
public class Demo {
    public static void main(String[] args) {
        Animal a1 = new Dog("Rex");
        Animal a2 = new Cat("Luna");
        Animal a3 = new Animal("Mystery"); // uses default speak()

        // Same call â†’ different results (polymorphism via overriding)
        System.out.println(a1.introduce()); // I'm Rex and I say: woof
        System.out.println(a2.introduce()); // I'm Luna and I say: meow
        System.out.println(a3.introduce()); // I'm Mystery and I say: ...

        // Works uniformly in collections
        Animal[] zoo = { a1, a2, a3 };
        for (Animal a : zoo) {
            System.out.println(a.getName() + " speaks: " + a.speak());
        }
    }
}
```

```java
Demo.main(null);
```
